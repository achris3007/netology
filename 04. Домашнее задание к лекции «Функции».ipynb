```python
# 1. перечень всех документов

documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

# 2.перечень полок, на которых хранятся документы (если документ есть в documents,
# то он обязательно должен быть и в directories)

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}



def call_function(command):
    if command == 'p':
        p()
    elif command == 's':
        s()
    elif command == 'l':
        l()
    elif command == 'ash':
        ash()
    elif command == 'dsh':
        dsh()
    elif command == 'ad':
        ad()
    elif command == 'd':
        d()
    elif command == 'm':
        m() 
    elif command == 'spisok':
        spisok()

        
# Служебные команды

def find_document_by_number(number):
    """
    Находит документ по номеру.
    """
    
    for line in documents:
        if number == line['number']:
            return line
        
    return None

def match():
    """
    Добавляет номер полки в перечень всех документов. 
    """
    d = {}
    
    for shelf, numbers in directories.items():
        for number in numbers:
            d[number] = shelf
        
    for line in documents:
        if line['number'] in d.keys():
            line['shelf'] = d[line['number']]

def find_shelf(shelf):
    """
    Сообщается, если полка существует.
    """
            
    for sh in directories.keys():
        if shelf == sh:
            return True
        
    return False     

def shelf_is_empty(shelf):
    """
    Проверяет отсутствие документов на полке.
    """
    if len(directories[shelf]) != 0:
        return False
        
    return True

def delete_doc_from_shelf(number):
    """
    Удаляет документ из полки.
    """
    
    line = find_document_by_number(number)
    match()
        
    for i, num in enumerate(directories[line['shelf']]):
            if line['number'] == num:
                del(directories[line['shelf']][i])
                break
    
    
# Пользовательские команды

def spisok():
    print('''
    p - узнать владельца документа по его номеру\n
    s - по номеру документа узнать на какой полке он хранится\n
    l - увидеть полную информацию по всем документам\n
    ash - добавить новую полку\n
    dsh - удалить полку\n
    ad - добавить новый документ в данные\n
    d - удалить документ из данных\n
    m - переместить документ с полки на полку\n
    q - завершить программу\n
    ''')
            
def p():
    """
    Пользователь по команде “p” может узнать владельца документа по его номеру.
    """
    user_input = input('Введите номер документа: ')
    line = find_document_by_number(user_input)
    
    if line is None:
        print('Документ не найден.')
    else: 
        print(line['name'])
        
def s():
    """
    Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится.
    """
    user_input = input('Введите номер документа: ')
    
    output = 'Документ не найден.'
    
    for shelf, number in directories.items():
        if user_input in number: 
            output = f'Документ хранится на полке: {shelf}'
    
    print(output)
                
def l():
    """
    Пользователь по команде “l” может увидеть полную информацию по всем документам.
    """
    match()
    
    print('Текущий список документов:')
    
    for line in documents:
        print(f"№: {line['number']}, тип: {line['type']}, владелец: {line['name']}, полка хранения: {line['shelf']}")
        
def ash():
    """
    Пользователь по команде “ash” может добавить новую полку.
    """
    user_input = input('Введите номер полки: ')
    shelf_exist = find_shelf(user_input)
    
    if shelf_exist is True:
        print(f'Такая полка уже существует. Текущий перечень полок: {list(directories.keys())}')
    else: 
        directories[user_input] = []
        print(f'Полка добавлена. Текущий перечень полок: {list(directories.keys())}')
        
def dsh(shelf = None):
    """
    Пользователь по команде “dsh” может удалить существующую полку из данных (только если она пустая).
    """
    
    if shelf is None:
        user_input = input('Введите номер полки: ')
    else:
        user_input = shelf
    
    if not find_shelf(user_input):
        output = f'Такой полки не существует. Текущий перечень полок: {list(directories.keys())}'
    elif shelf_is_empty(user_input) == True:
        del(directories[user_input])
        output = f'Полка удалена. Текущий перечень полок: {list(directories.keys())}' 
    elif shelf_is_empty(user_input) == False: 
        output = f'На полке есть документ(ы), удалите их перед удалением полки. Текущий перечень полок: {list(directories.keys())}'
    
    print(output)

def ad():
    """
    Пользователь по команде “ad” может добавить новый документ в данные.
    """
    user_input_number = input('Введите номер документа: ')
    user_input_type = input('Введите тип документа: ')
    user_input_name = input('Введите владельца документа: ')
    user_input_shelf = input('Введите полку для хранения: ')
    
    add_shelf = find_shelf(user_input_shelf)
    
    if add_shelf is False:
        print('Такой полки не существует. Добавьте полку командой ash.')
        return
    elif add_shelf is True:
        directories[user_input_shelf].append(user_input_number)
        
    line = find_document_by_number(user_input_number)
    
    if line is None:
        new_line = {}
        new_line['number'] = user_input_number
        new_line['type'] = user_input_type
        new_line['name'] = user_input_name
        documents.append(new_line)
    else:
        print('Документ уже есть в базе.')
    
    l()
    
def d(number = None):
    """
    Пользователь по команде “d” может удалить документ из данных.
    """
    if number is None:
        user_input = input('Введите номер документа: ')
    else:
        user_input = number
        
    line = find_document_by_number(user_input)
    
    if line is None:
        print('Документ не найден.')
    else:
        match()
        delete_doc_from_shelf(line['number'])
        for i, el in enumerate(documents):
            if line['number'] == el['number']:
                del(documents[i])
                break
    
    l()
    
def m():
    """
    Пользователь по команде “m” может переместить документ с полки на полку.
    """
    
    user_input_number = input('Введите номер документа: ')
    user_input_shelf = input('Введите полку для хранения: ')
    
    line = find_document_by_number(user_input_number)
    
    if line is None:
        print('Документ не найден.')
    else:
        user_shelf = find_shelf(user_input_shelf)
        if user_shelf is False:
            print(f'Такой полки не существует. Текущий перечень полок: {list(directories.keys())}')
        else:
            delete_doc_from_shelf(user_input_number)
            directories[user_input_shelf].append(user_input_number)
            match()
            print('Документ перемещен.')
            
    l()

    
    
command = ""
while command != 'q':
    command = input('Введите команду в поле ниже. Введите "spisok", чтобы увидеть список доступных команд.\n')
    call_function(command)
else:
    print('Программа завершена')
```
