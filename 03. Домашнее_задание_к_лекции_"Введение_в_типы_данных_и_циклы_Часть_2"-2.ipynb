## Задание 2

```python
queries = [
    'смотреть сериалы онлайн бесплатно сейчас',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]

d = {}

for phrase in queries:
    splitted_phrase = phrase.split(' ')
    l = len(splitted_phrase)
    if l not in d:
        d[l] = 1 
    else:
        d[l] += 1

for words, count in sorted(d.items()):
    print(f'Поисковых запросов, содержащих {words} слов(в): {count / len(queries) * 100:0.2f}%')

# x = 0
# y = 0
# x += 1
# y += 1
# print(f'Поисковых запросов, содержащих {x} слов(в): {x / len(queries) * 100:0.2f}%')
# # print(f'Поисковых запросов, содержащих {y} слов(в): {y / len(queries) * 100:0.2f}%')
```

## Задание 3

```python
results = {
    'vk': {'revenue': 103, 'cost': 98},
    'yandex': {'revenue': 179, 'cost': 153},
    'facebook': {'revenue': 103, 'cost': 110},
    'adwords': {'revenue': 35, 'cost': 34},
    'twitter': {'revenue': 11, 'cost': 24},
}
 
print('Результат:\n')

for media, info in sorted(results.items()):
    info['ROI'] = (info['revenue'] / info['cost'] - 1) * 100
    print(f'{media}: revenue: {info["revenue"]}, cost: {info["cost"]}, ROI: {info["ROI"]:0.2f}')
    
# ROI = (revenue / cost - 1) * 100

# print(f'Результат:\n{sorted(results.items())}')
```

# Задание 5

```python
my_list = ['a', 'b', 'c', 'd', 'e', 'f']

my_list_reverse = my_list[::-1]
last_el = my_list_reverse[0]
d = {}

for el in my_list_reverse[1:]:
    d = {el: last_el}
    last_el = d

print(f'Результат: {d}')
```
